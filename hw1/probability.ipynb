To find $P(M \cap W)$ (Probability UVA wins both games), we use:

Substituting the given values:

$P(M \cap W) = 0.60 + 0.60 - 0.75 = 0.45$

## 1(c):

- $P(W) = 0.60$ (Probability UVA women win)

$P(M | W) = \frac{P(M \cap W)}{P(W)} = \frac{0.45}{0.60} = 0.75$

The probability of that the UVA men's team wins, give the women's team has won is $75\%$.

## 2: 

To calculate the probability of picking two matching socks from the drawer, we start by determining the total number of ways to pick any two socks and the number of ways to pick two matching socks

Using the combination formula with $n=8$ (total socks) and $k=2$ (socks to pick), we have:

$\text{Total Ways to Pick 2} = \frac{n!}{k!(n-k)!} = \frac{8!}{2!(8-2)!} = \frac{8 \times 7}{2} = 28$

Since there are 4 pairs of socks, and you can only pick a matching pair from the same color, there are 4 ways to pick matching socks (one way per color).

The probability is the ratio of the number of ways to pick two matching socks to the total number of ways to pick any two socks:

$$\text{Probability} = \frac{\text{Number of Ways to Pick Matching Socks}}{\text{Total Ways}} = \frac{4}{28}$$

The probability that you pick two socks that match is $\frac{1}{7}$ or $0.143$.

## 3(a):

- Let L be lung cancer positive \
- Let S be people who are smokers

To solve this problem, we calculate $P(S \cap L)$ and $P(S \cap \neg L)$, and simply sum them up to get the final result.

The probability of being a smoker and having lung cancer can be calculated as:
$P(S \cap L) = P(S|L) \times P(L)$

The probability of being a smoker and not having lung cancer can be calculated as:
$P(S \cap \neg L) = P(S|\neg L) \times P(\neg L)$
where:
$P(\neg L) = 1 - P(L)$

The total percentage of the population that are smokers is the sum of the above probabilities:
$P(S) = P(S \cap L) + P(S \cap \neg L)$

Substituting the given values:

- $P(L) = 0.0013$
- $P(S|L) = 0.9$
- $P(S|\neg L) = 0.169$

We calculate:

1. $P(S \cap L) = 0.9 \times 0.0013 = 0.00117$
2. $P(S \cap \neg L) = 0.169 \times (1 - 0.0013) = 0.16878$
3. Therefore, the total percentage of the population that are smokers is approximately $P(S) = 0.00117 + 0.16878 = 16.995\%$.



## 3(b): 

To determine the probability of developing lung cancer given that one is a smoker, \(P(L|S)\), we can apply the conditional probability formula with values previously calculated:
$P(L|S) = \frac{P(L \cap S)}{P(S)}$

### Recap of Calculated Values:

- We previously calculated $P(L \cap S)$, the probability of being a smoker and having lung cancer, as 0.00117.
- We also determined $P(S)$, the probability of being a smoker within the total population, to be approximately 16.995% or 0.16995.

### Applying the Formula:

Given the above, we can directly substitute into the formula:
$P(L|S) = \frac{0.00117}{0.16995}$

So the probability of developing lung cancer if you are smoker is 0.00688 or $0.688\%$.




## 3(C):


To find the probability of developing lung cancer for individuals who are not smokers, we utilize the conditional probability formula:
$P(L|S^C) = \frac{P(L \cap S^C)}{P(S^C)}$

- The probability of being a smoker, $P(S) = 0.17$, was calculated earlier.
- The probability of developing lung cancer, $P(L) = 0.0013$.


1. **Calculate $P(S^C)$**: The probability of not being a smoker.
   - $P(S^C) = 1 - P(S) = 1 - 0.17 = 0.83$

2. **Calculate $P(L \cap S^C)$**: The probability of developing lung cancer while not being a smoker.
   - We previously calculated $P(L \cap S) = 0.00117$.
   - Therefore, $P(L \cap S^C) = P(L) - P(L \cap S) = 0.0013 - 0.00117 = 0.00013$

3. **Complete the Calculation**:
   - Substituting the values into the formula gives us:
   $$P(L|S^C) = \frac{0.00013}{0.83}$$


Performing the calculation provides the probability of a non-smoker developing lung cancer:
- $P(L|S^C) = \frac{0.00013}{0.83} \approx 0.0001566$

So the probability of developing lung cancer if you are not smoker is 0.0001566 or $0.0157\%$.

```python
## 4(c):

import numpy as np

# Step 1: Pick a prior by getting frequency of classes in training data
training_data = input[input['TrainData'] == 1]
test_data = input[input['TrainData'] == 0]
total_len = len(training_data)

dementia_freq = len(training_data[training_data['Dementia'] == 1])
healthy_freq = len(training_data[training_data['Dementia'] == 0])

prob_healthy = healthy_freq / total_len
prob_dementia = dementia_freq / total_len

# Step 2: Train you Gaussians on training data, finding mean and variance 

healthy_train = training_data[training_data['Dementia'] == 0]
dementia_train = training_data[training_data['Dementia'] == 1]

mean_right_healthy = healthy_train['RightHippoVol'].mean()
var_right_healthy = healthy_train['RightHippoVol'].var()

mean_left_healthy = healthy_train['LeftHippoVol'].mean()
var_left_healthy = healthy_train['LeftHippoVol'].var()

mean_right_dementia = dementia_train['RightHippoVol'].mean()
var_right_dementia = dementia_train['RightHippoVol'].var()

mean_left_dementia = dementia_train['LeftHippoVol'].mean()
var_left_dementia= dementia_train['LeftHippoVol'].var()


def calculate_probability(x, mean, var):
    exponent = np.exp(-((x-mean)**2 / (2 * var)))
    return (1 / np.sqrt(2 * np.pi * var)) * exponent

counter = 0

for index, row in test_data.iterrows():
    prob_density_right_healthy = calculate_probability(row['RightHippoVol'], mean_right_healthy, var_right_healthy)
    prob_density_left_healthy = calculate_probability(row['LeftHippoVol'], mean_left_healthy, var_left_healthy)
    prob_density_right_dementia = calculate_probability(row['RightHippoVol'], mean_right_dementia, var_right_dementia)
    prob_density_left_dementia = calculate_probability(row['LeftHippoVol'], mean_left_dementia, var_left_dementia)
    
    
    posterior_healthy = prob_density_left_healthy * prob_density_right_healthy
    posterior_dementia = prob_density_right_dementia * prob_density_left_dementia
    
    prob_evidence = posterior_healthy * prob_healthy + posterior_dementia * prob_dementia
    
    final_healthy_prob = (posterior_healthy * prob_healthy) / prob_evidence
    final_dementia_prob = (posterior_dementia * prob_dementia) / prob_evidence
    if final_dementia_prob > 0.5:
        prediction = 1
    else:
        prediction = 0
    
    if prediction == row['Dementia']:
        counter += 1

classifier_accuracy = counter / len(test_data)
print(classifier_accuracy)
```

The accuracy is about 71% using frequency of classes in training data as a prior.
